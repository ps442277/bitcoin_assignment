# Pandas vs Pandas API commands

Pandas API has a similar syntax to pandas (native pandas), and no spark knowledge is required to work with it. The below commands compare the pandas commands to Pandas API (Pandas-on-spark).

```pyspark
#Pandas series:
nat_series = pd.Series([1,2,3,4,np.nan])
nat_series
```

```pyspark
# Pandas API series: 
panAPI_series = ps.Series([1, 3, 5, np.nan, 6, 8])
panAPI_series
```

The above two commands show that pandas and pandas API have similar syntax.

```pyspark
#creating native Pandas dataframe. The basic syntax is the same in both commands.

nat_dataframe = pd.DataFrame(
    {'a': [1, 2, 3, 4, 5, 6],
     'b': [100, 200, 300, 400, 500, 600],
     'c': ["one", "two", "three", "four", "five", "six"]},
    index=[10, 20, 30, 40, 50, 60])

nat_dataframe
```

```pyspark
# In case of Pandas API, Spark job is created which is not present in case of pandas. 

panAPI_dataframe = ps.DataFrame(
    {'a': [1, 2, 3, 4, 5, 6],
     'b': [100, 200, 300, 400, 500, 600],
     'c': ["one", "two", "three", "four", "five", "six"]},
    index=[10, 20, 30, 40, 50, 60])

panAPI_dataframe
```

```pyspark
#get summarized results on pandas-spark-dataframe on numerical columns
panAPI_dataframe.describe()
```

```pyspark
#Property of Type Support: Existing pandas DataFrame can be converted to a pandas-on-Spark DataFrame.
# using above nat_dataframe. 

panAPI_datframe_new = ps.from_pandas(nat_dataframe)
type(panAPI_datframe_new)
```

# Working with Spark vs. Pandas API commands

To use the spark framework, you need to learn Spark SQL, but Pandas API enables python users to run pandas on spark.

```pyspark
#Property of Type Support: Existing pandas DataFrame can be converted to a Spark DataFrame.
#It is also a spark job.

spark = SparkSession.builder.getOrCreate()
spark_dataframe = spark.createDataFrame(nat_dataframe)
spark_dataframe.show()
```

```pyspark
#Property of Type Support: Existing Spark DataFrame can be converted to native pandas DataFrame.
spark_topandas_dataframe = spark_dataframe.toPandas()
type(spark_topandas_dataframe)
```

```pyspark
#Property of Type Support: Existing native pandas DataFrame can be converted to pandas-on-spark DataFrame.
pandasAPI_datframe = ps.from_pandas(spark_topandas_dataframe)
type(pandasAPI_datframe)
```

Information:
- pandasAPI_datframe is pandas-API-dataframe
- spark_topandas_dataframe is spark dataframe
- nat_dataframe is native panda dataframe

# SQL queries on top of pandas-on-spark

One of the properties of Pandas-on-Spark is that it can run SQL queries on top of it. This simplifies the querying process as the learners have exposure to writing SQL commands.

```pyspark
#sql query for selecting data in 2021

ps.sql("select * from {pandas_api_import} where Year in ('2021')")
```

```pyspark
#sql query for selecting data in 2021 and Industry_aggregation_NZSIOC in Level 3.

ps.sql("select * from {pandas_api_import} where Year in ('2021') and Industry_aggregation_NZSIOC in ('Level 3')")
```

```pyspark
#sql select query for selecting data year,Industry_aggregation_NZSIOC columns  in 2021 and Industry_aggregation_NZSIOC in Level 3.

ps.sql("select year,Industry_aggregation_NZSIOC from {pandas_api_import} where Year in ('2021') and Industry_aggregation_NZSIOC in ('Level 3')")
```

```pyspark
#query to get the count of Level group by year.

ps.sql("select count(Industry_aggregation_NZSIOC), Year from {pandas_api_import} GROUP BY Year")
```

# Performance Comparison. 

We will use the describe function to see the difference in Native Pandas vs. Pandas-API. In terms of memory, when you try to run the below command with range larger range like 100000000, it will cause the Native pandas describe function to fail due to insufficient memory on single node. But when you will increase the memory, the native pandas will consume more time than Pandas API.

```pyspark
dataframe = spark.range (100000)
dataframe.show()
```

```pyspark
#converting spark dataframe to pandas-on-spark dataframe
pandas_on_spark_import = dataframe.to_pandas_on_spark()
pandas_on_spark_import
```

```pyspark
#converting spark dataframe to pandas dataframe
new_native_pandas_import = dataframe.toPandas()
new_native_pandas_import
```

```pyspark
#running descibe function on Pandas API dataframe
pandas_on_spark_import.describe()
```

```pyspark
#running descibe function on native Pandas dataframe
new_native_pandas_import.describe()
```
